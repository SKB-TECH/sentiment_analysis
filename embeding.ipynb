```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=['the glass of milk',
'the glass of juice',
'the cup of tea',
'I am a good boy',
'I am a good developer',
'understand the meaning of words',
'your videos are good'
]
```

```python
## define vocabulaire size
voc_size=10000
```

```python
## one hot representation
one_hot_representation=[one_hot(words,voc_size) for words in sent]
one_hot_representation
```

```python
## word Embeding representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length=8
embedding_docs=pad_sequences(one_hot_representation,padding='pre',maxlen=sent_length)
```

```python
print(embedding_docs)
```

```python
dim=10
```

```python
## feature-representation
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.predict(embedding_docs)
```
